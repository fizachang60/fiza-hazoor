```python
import pandas as pd
import matplotlib.pyplot as plt

### SIMPLE DATA


data = {
    "Student": ["Ali", "Sara", "Ahmed", "Ayesha"],
    "Marks": [80, 90, 75, 95]
}

df = pd.DataFrame(data)
print(df)

### BAR CHART
plt.bar(df["Student"], df["Marks"])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

### LINE CHART
plt.plot(df["Student"], df["Marks"], marker="o")
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
