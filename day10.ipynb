```python
import pandas as pd

# Create Data
sales_data = {
    "product": ["Laptop", "Phone", "Mouse", "Charger", "Tablet"],
    "Sales": [900, 700, 400, 120, 450]
}

# Create DataFrame
df = pd.DataFrame(sales_data)

# Show DataFrame
print(df)

# Export DataFrame to CSV
df.to_csv("sales_data.csv", index=False)

# Success Message
print("Data exported successfully!")
print("File Name: sales_data.csv")
```
