```python
import pandas as pd
data  = {
    "name":["sobia", "aqsa", "fiza"],
    "age": [20, 19 ,18],
    "marks":[80, 70,  90]
}
df = pd.DataFrame(data)
print(df)

## MISSING VALUE CHECK
print(df.isnull())
print(df.isnull().sum())

### MISSING VALUE FILL
df["marks"] = df["marks"].fillna(0)
df["name"] = df["name"].fillna("unknown")
print(df)

### DUPLICATE ROW REMOVE
df = df.drop_duplicates()
print(df)

#### DATA TYPEs
print("data types:")
print(df.dtypes)

#####FINAL CLEAN DATA TYPES
print("cleaned data:")
print(df)
```
